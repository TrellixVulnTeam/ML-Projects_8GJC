## Matplotlib Example Workflow

```python
# 0 importmathplotlib and get ready for plotting jupyer

%matplotlib inline 
import matplotlib.pyplot as plot 

# some data
x = [2,4,6,8]
y = [11,22,33,44]

# setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3 plot the data
ax.plot(x,y)

#4 customize plot 
ax.set(title="Simple Plot", 
      xlabel="x-axis",
      ylabel="y-axis")

# 5 save and show (you save the whole figure)

fig.savefig("images/sample-plot.png")
```

## Making figures with Numpy array

```python
import numpy as np
```

```python
# create some data

x = np.linspace(0,10,100)
x[:20]
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# use some data to make scatter

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# make another scatter

fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Make plot from dictionary

videocard_prices = {"3060 Ti" : 500,
                    "3070": 600,
                   "3080" : 800,
                   "3090": 1500}
fig, ax = plot.subplots()
ax.bar(videocard_prices.keys(), videocard_prices.values())
ax.set(title="Video Card Prices", ylabel="Prices ($)");
```

```python
fig, ax = plot.subplots()
ax.barh(list(videocard_prices.keys()), list(videocard_prices.values()));
```

```python
# make hist and plot it
# make some data
# create a normal dis
x = np.random.randn(1000)
fig, ax = plot.subplots()
ax.hist(x);
```

```python
# subplot option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# plot to different axis 
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(videocard_prices.keys(), videocard_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# SUbplots options 2

fig, ax = plt.subplots(nrows=2,
                      ncols=2, 
                      figsize=(10,5))
# plots to reach different index
# use index to position the figuures 

ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(50), np.random.random(50));
ax[1,0].bar(videocard_prices.keys(), videocard_prices.values());
ax[1,1].hist(np.random.random(1000));
```

## Plotting from pandas DataFrame

```python
import pandas as pd
```

```python
# make dataframe

car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2010", periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
type(car_sales["Price"][0])
```

```python
# remove last two zeros

car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
 car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
 car_sales   
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
type(car_sales["Price"][0])
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Plots total sale over time

car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# convert price column to int first 
car_sales["Price"] = car_sales["Price"].astype(int)

# scatter plot with price as numeric
car_sales.plot (x="Odometer (KM)", y="Price", kind="scatter");
```
